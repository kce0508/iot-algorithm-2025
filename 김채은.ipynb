2. P269 유명 맛집 대기줄 구현하기 문제를 while True:문을 사용하여 동적으로 입력을 받는 프로그램으로 변경작성하시오.

```python
# 함수 선언
def isQueueFull():
    global SIZE, queue, rear
    return rear == SIZE - 1  # rear가 큐의 마지막 인덱스에 도달하면 큐가 꽉 찼다고 판단

def isQueueEmpty():
    global front, rear
    return front == rear  # front와 rear가 같으면 큐가 비어 있다고 판단

def enQueue(data):
    global SIZE, queue, rear
    if isQueueFull():
        print('큐가 꽉 찼습니다.')
        return
    rear += 1
    queue[rear] = data  # 새로운 데이터를 rear 위치에 삽입
    print(f'✅ {data} 님이 대기 줄에 추가되었습니다.')

def deQueue():
    global SIZE, queue, front, rear
    if isQueueEmpty():
        print('큐가 비었습니다.')
        return None
    front += 1
    data = queue[front]  # front 위치에 있는 데이터 반환
    queue[front] = None  # 제거된 자리 비우기

    # 모든 사람을 한 칸씩 앞으로 이동
    for i in range(front + 1, rear + 1):
        queue[i - 1] = queue[i]
        queue[i] = None
    rear -= 1  # rear도 한 칸 감소

    return f' {data} 님 식당에 입장!'

def peek():
    global SIZE, queue, front, rear
    if isQueueEmpty():
        print('큐가 비었습니다.')
        return None
    return queue[front + 1]

# 전역 변수 선언
SIZE = 5
queue = [None] * SIZE
front = rear = -1  # 초기 상태는 비어있음

# 동적으로 입력 받는 메인 코드
if __name__ == '__main__':
    while True:
        print('\n=== 대기 줄 관리 시스템 ===')
        print('1. 대기 줄 추가 (enQueue)')
        print('2. 식당 입장 (deQueue)')
        print('3. 현재 대기 줄 보기')
        print('4. 프로그램 종료 (exit)')

        command = input('메뉴를 선택하세요: ')

        if command == '1':
            name = input('이름을 입력하세요: ')
            enQueue(name)

        elif command == '2':
            print(deQueue())

        elif command == '3':
            print('현재 대기 줄 상태: ', queue)

        elif command == '4' or command.lower() == 'exit':
            print('식당 영업 종료!')
            break

        else:
            print('잘못된 입력입니다. 다시 선택하세요.')
```

3. P309 편의점 판매된 물건 목록 출력하기 문제 역시 while True: 문으로 동적으로 입력받는 프로그램으로 변경하시오.
데이터의 종류는 [ 콘말차, 삿포로 맥주, 아카페라 벤티 헤이즐넛, 레어치즈푸딩, 척오리지널 메가 사워, 요아정 요거트컵, 페퍼로니피자주먹밥, 널담 슈톨렌, 딸기마시멜로케이크, 버니공쥬 딸기뚱카롱, 고추잡채호빵, 체다 슈크림붕어스낵 ]

```python
import random
# 함수 선언
class TreeNode():
    def __init__(self):
        self.left = None
        self.data = None
        self.right = None

# 전역 변수 선언
memory = []
root = None
dataAry = ['콘말차', '삿포로 맥주', '아카페라 벤티 헤이즐넛', '레어치즈푸딩', '척오리지널 메가 사워', '요아정 요거트컵', '페퍼로니피자주먹밥', 
           '널담 슈톨렌', '딸기마시멜로케이크', '버니공쥬 딸기뚱카롱', '고추잡채호빵', '체다 슈크림붕어스낵']
sellAry = []

# 동적으로 입력 받는 메인코드
print("오늘 판매된 물건을 입력하세요.(입력을 마치려면 'exit'입력)")

while True:
    item = input("판매된 물건을 입력하세요: ")
    if item.lower() == "exit":
        break
    elif item in dataAry:   # 목록에 있으면 리스트에 추가
        sellAry.append(item)
    else:
        print("입력한 물건이 목록에 없습니다. 목록에서 선택해 주세요.")

print(f'오늘 판매된 물건(중복O) --> {sellAry}')

# 메인 코드
node = TreeNode()
node.data = sellAry[0]
root = node
memory.append(node)

# 나머지 판매된 물건들을 트리에 삽입
for name in sellAry[1:]:

    node = TreeNode()
    node.data = name

    current = root
    while True:
        if name == current.data:
            break
        if name < current.data:
            if current.left == None:
                current.left = node
                memory.append(node)
                break
            current = current.left
        else:
            if current.right == None:
                current.right = node
                memory.append(node)
                break
            current = current.right

print('이진 탐색 트리 구성 완료!')

# 이진 탐색 트리 전위 순회 함수(preorder)
def preorder(node):
    if node == None:
        return
    print(node.data, end=' ')
    preorder(node.left)
    preorder(node.right)

print('오늘 판매된 종류(중복X)--> ', end=' ')
preorder(root)
```

4. P473 Self Study 13-2 의 문제. 이진 검색 알고리즘을 풀면서 검색 횟수도 같이 출력하도록 구현하세요.

```python
import random

# 이진 탐색 함수
def binSearch(ary, fData):
    pos = -1
    start = 0
    end = len(ary) - 1
    count = 0

    while(start <= end):
        count += 1  # 탐색 시마다 횟수 증가
        mid = (start + end) // 2
        if fData == ary[mid]:
            return mid, count   # 검색 성공시 위치와 횟수 반환
        elif fData > ary[mid]:
            start = mid + 1
        else:
            end = mid - 1

    return pos, count

# 랜덤한 숫자 배열 생성
dataAry = [random.randint(0, 99999) for _ in range(100000)]  # 0~100000 사이의 숫자 100000개
sortedAry = sorted(dataAry)  # 정렬

# 검색을 시도할 데이터 생성
findData = random.randint(0, 100000)

# 배열 일부 출력
print(f'배열 일부 --> {sortedAry[:5]} ~ {sortedAry[-5:]}')

# 이진 검색
pos, count = binSearch(sortedAry, findData)

if pos != -1:
    print(f'{findData}(은)는 {pos} 위치에 있음')
else:
    print(f'{findData}(이)가 없네요.')
    
print(f'## {count}회 검색함')
```
